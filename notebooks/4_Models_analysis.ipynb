Для feature extraction MobileNetV2 показла погані результати, але для fine_tuning в одному випадку буда найгіршою, в іншому - одна з кращих. ConvNeXtTiny, NASNetMobile, InceptionV3 внизу таблиці в обох випадках.

```python
# візуалізація
def plot_scores(df, metrics_list, title='', sorted=False, sortby=None, ylim=None):
  n = len(metrics_list)
  sortby = metrics_list[0] if sortby is None else sortby
  # palette = sns.color_palette("husl", len(metrics_list))
  if n == 1:
    plt.figure(figsize=(10, 5))
    plt.ticklabel_format(style='plain', axis='y')
    if sorted:
      df = df.sort_values(by=sortby, ascending=False)
    sns.barplot(x=df['Model'], y=df[metrics_list[0]], palette='husl', hue=df['Model'], legend=False)
    plt.title(f'{metrics_list[0]} {title}')
    plt.xticks(rotation=15, ha="right")
    if ylim is not None:
      plt.ylim(ylim[0], ylim[1])

  else:
    rows = int(np.ceil(n/2))
    cols = 2
    fig, axes = plt.subplots(rows, cols, figsize=(18, rows * 5))
    fig.suptitle(f'Порівняння моделей, що навчалися на вибірці, {title}')
    for metric, ax in zip(metrics_list, axes.flatten()):
      if sorted:
        df = df.sort_values(by=sortby, ascending=False)
      sns.barplot(x=df['Model'], y=df[metric], ax=ax, palette='husl', hue=df['Model'], legend=False)
      ax.set_title(metric)
      ax.set_xticklabels(df['Model'], ha="right", rotation=20)
      ax.grid(True)
      if ylim is not None:
        ax.set_ylim(ylim[0], ylim[1])
  plt.show()
```

```python
def plot_scores_with_labels(df, metrics_list, title='', sorted=False, sortby=None, ylim=None):
    n = len(metrics_list)
    sortby = metrics_list[0] if sortby is None else sortby

    if n == 1:
        plt.figure(figsize=(10, 5))
        plt.ticklabel_format(style='plain', axis='y')
        if sorted:
            df = df.sort_values(by=sortby, ascending=False)

        ax = sns.barplot(x=df['Model'], y=df[metrics_list[0]], palette='husl', hue=df['Model'], legend=False)
        plt.title(f'{metrics_list[0]} {title}')
        plt.xticks(rotation=15, ha="right")
        if ylim is not None:
            plt.ylim(ylim[0], ylim[1])

        # Мітки
        for p in ax.patches:
            ax.annotate(f'{p.get_height():.4f}',
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='bottom', fontsize=10, color='black', fontweight='bold')

    else:
        rows = int(np.ceil(n / 2))
        cols = 2
        fig, axes = plt.subplots(rows, cols, figsize=(18, rows * 5))
        fig.suptitle(f'Порівняння моделей, що навчалися на вибірці, {title}')

        for metric, ax in zip(metrics_list, axes.flatten()):
            if sorted:
                df = df.sort_values(by=sortby, ascending=False)

            sns.barplot(x=df['Model'], y=df[metric], ax=ax, palette='husl', hue=df['Model'], legend=False)
            ax.set_title(metric)
            ax.set_xticklabels(df['Model'], ha="right", rotation=20)
            ax.grid(True)
            if ylim is not None:
                ax.set_ylim(ylim[0], ylim[1])

            # Мітки
            for p in ax.patches:
                ax.annotate(f'{p.get_height():.4f}',
                            (p.get_x() + p.get_width() / 2., p.get_height()),
                            ha='center', va='bottom', fontsize=10, color='black', fontweight='bold')

    plt.show()
```

```python
def models_scatter(df, test_dataset_name, metrics_names=['precision', 'recall'], title_str='different Models', ylim=None, xlim=None):
  plt.figure(figsize=(6, 4))

  colors = plt.cm.tab10(range(df.shape[0]))

  for model, prec, rec, color in zip(
      df['Model'],
      df[metrics_names[0]],
      df[metrics_names[1]],
      colors):
      plt.scatter(prec, rec, color=color, label=model, s=100, edgecolors='black')

  plt.xlabel(metrics_names[0], fontsize=14)
  plt.ylabel(metrics_names[1], fontsize=14)
  plt.title(f"{metrics_names[0]} vs {metrics_names[1]} for {title_str} on {test_dataset_name} data")
  plt.grid(True)
  plt.legend(title="Models", fontsize=12, loc="upper left", bbox_to_anchor=(1, 1))
  if ylim is not None:
      plt.ylim(ylim[0], ylim[1])
  if xlim is not None:
      plt.xlim(xlim[0], xlim[1])

  plt.show()
```

```python
models_scatter(dfs[0], 'test 2020', ylim=[0.1, 0.7], xlim=[0.1, 0.7])
models_scatter(dfs[1], 'cropped test 2020', ylim=[0.1, 0.7], xlim=[0.1, 0.7])
```

```python
ensembles_df = pd.read_excel('Compare_ensembles.xlsx', sheet_name='ensembles')
ensembles_df.sort_values(by='auc', ascending=False)
```

```python
def models_scatter_ensemble(df, test_dataset_name, metrics_names=['precision', 'recall'], title_str='different Models', ylim=None, xlim=None):
  plt.figure(figsize=(6, 4))

  colors = plt.cm.tab10(range(df.shape[0]))

  for model, prec, rec, color in zip(
      df['Meta-model'],
      df[metrics_names[0]],
      df[metrics_names[1]],
      colors):
      plt.scatter(prec, rec, color=color, label=model, s=100, edgecolors='black')

  plt.xlabel(metrics_names[0], fontsize=14)
  plt.ylabel(metrics_names[1], fontsize=14)
  plt.title(f"{metrics_names[0]} vs {metrics_names[1]} for {title_str} on {test_dataset_name} data")
  plt.grid(True)
  plt.legend(title="Models", fontsize=12, loc="upper left", bbox_to_anchor=(1, 1))
  if ylim is not None:
      plt.ylim(ylim[0], ylim[1])
  if xlim is not None:
      plt.xlim(xlim[0], xlim[1])

  plt.show()
```

```python
models_scatter_ensemble(ensembles_df, 'test 2019+2020', title_str='Ensebmles', ylim=[0.55, 0.7], xlim=[0.4, 0.5])
```
